```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
#Creating the dataset 
df = pd.read_csv(r"Absenteeism_at_work.csv" , delimiter=';')
print(df.head())
```

```python
#Cleanig the columns 
df.columns = df.columns.str.strip()
```

```python
#selecting relevant features 
selected_features = [
     "Age", "Distance from Residence to Work", "Transportation expense",
    "Service time", "Work load Average/day", "Body mass index", "Absenteeism time in hours"
]
```

```python
print(df.columns.tolist())
```

```python
df.columns = df.columns.str.strip()  # Removes spaces from column names
df["Absenteeism Category"] = pd.qcut(df["Absenteeism time in hours"], q=3, labels=["Low", "Medium", "High"])
```

## Creating pairplot for "MUltivariate distribution".

```python
#Creaating pairplot for multivariate distribution 
sns.pairplot(df[selected_features + ["Absenteeism Category"]],
             diag_kind= "kde", corner= True, hue = "Absenteeism Category", palette= 'coolwarm') 

plt.show()
```
