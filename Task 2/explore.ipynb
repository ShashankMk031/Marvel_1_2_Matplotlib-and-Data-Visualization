```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets 
%matplotlib inline 

# Load Iris dataset
iris = sns.load_dataset("iris")

# Set style
sns.set_style("whitegrid")

# Create a figure with multiple plots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Line Plot
x = np.linspace(0, 10, 100)
y = np.sin(x)
axes[0, 0].plot(x, y, label='Sine Wave')
axes[0, 0].set_title("Line Plot")
axes[0, 0].set_xlabel("X-axis")
axes[0, 0].set_ylabel("Y-axis")
axes[0, 0].legend()

# Scatter Plot using Iris dataset
sns.scatterplot(data=iris, x="sepal_length", y="sepal_width", hue="species", ax=axes[0, 1])
axes[0, 1].set_title("Scatter Plot")

# Bar Plot
sns.barplot(x=["A", "B", "C"], y=[10, 20, 15], ax=axes[1, 0])
axes[1, 0].set_title("Bar Plot")

# Histogram
sns.histplot(iris["petal_length"], bins=20, kde=True, ax=axes[1, 1])
axes[1, 1].set_title("Histogram")

plt.tight_layout()
plt.savefig("multiple_plots.png")
plt.show()

# Additional Plots
plt.figure(figsize=(6, 6))
labels = ["A", "B", "C", "D"]
values = [30, 10, 40, 20]
plt.pie(values, labels=labels, autopct='%1.1f%%', colors=['blue', 'red', 'green', 'purple'])
plt.title("Pie Chart")
plt.savefig("pie_chart.png")
plt.show()

# Box Plot
sns.boxplot(data=iris, x="species", y="petal_length")
plt.title("Box Plot")
plt.savefig("box_plot.png")
plt.show()

# Violin Plot
sns.violinplot(data=iris, x="species", y="petal_length")
plt.title("Violin Plot")
plt.savefig("violin_plot.png")
plt.show()

# Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(iris.select_dtypes(include=['float64', 'int64']).corr(), annot=True, cmap="coolwarm")
plt.title("Heatmap")
plt.savefig("heatmap.png")
plt.show()


# 3D Plot
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(iris["sepal_length"], iris["sepal_width"], iris["petal_length"], c=iris.index)
ax.set_xlabel("Sepal Length")
ax.set_ylabel("Sepal Width")
ax.set_zlabel("Petal Length")
plt.title("3D Scatter Plot")
plt.savefig("3d_plot.png")
plt.show()
```
